## Building the Polymerization model in GillesPy2
***
Constructing this model in GillesPy2 reqires that we specify a maximum $n$ and $m$ that we are keeping track of. $n$, $m$, and $n+m$ must all be less than the maximum signed integer ($2^{31}-1$).  Then we create a new Species object for each state: $P_1$, $P_2$, ... $P_n$, and $D_1$, $D_2$, ... $D_{n+m}$.

```python
def build_styrene_polymerization(MAX_N=10):
    # Initialize Model
    model = gillespy2.Model(name="StyrenePolymerization")

    # First, choose a maximum value for the states
    max_state = MAX_N
    print(f"Creating Styrene Polymerization model, MAX_N={MAX_N}")
    
    # Define Variables (GillesPy2.Species)
    I = gillespy2.Species(name="I", initial_value=100, mode="discrete")
    R = gillespy2.Species(name="R", initial_value=0, mode="discrete")
    M = gillespy2.Species(name="M", initial_value=100, mode="discrete")
    species = [I, R, M]
    
    for n in range(1, max_state):
        species.append(gillespy2.Species(name=f"P{n}", initial_value=0, mode="discrete"))
        species.append(gillespy2.Species(name=f"D{n}", initial_value=0, mode="discrete"))
        
    # Add Variables to Model
    model.add_species(species)

    # Define Parameters
    k_d = gillespy2.Parameter(name="k_d", expression="1")
    k_i = gillespy2.Parameter(name="k_i", expression="1")
    k_iterm = gillespy2.Parameter(name="k_iterm", expression="1")
    k_p = gillespy2.Parameter(name="k_p", expression="1")
    k_tm = gillespy2.Parameter(name="k_tm", expression="1")
    
    # for the a=1/2
    k_tm_2 = gillespy2.Parameter(name="k_tm_2", expression="0.5")
    k_tc = gillespy2.Parameter(name="k_tc", expression="1")
    
    # Add Parameters to Model
    model.add_parameter([k_d, k_i, k_iterm, k_p, k_tm, k_tm_2, k_tc])

    # Define Initiation Reactions
    r1 = gillespy2.Reaction(reactants={'I': 1}, products={'R': 2}, rate="k_d")
    r2 = gillespy2.Reaction(reactants={'R': 1, 'M': 1}, products={'P1': 1}, rate="k_d")

    # Techically, you cannot have a trimer reaction is mass-action kinetics.
    # However, we can use a custom propensity to make it work.
    r3 = gillespy2.Reaction(reactants={'M': 3}, products={'R': 2}, propensity_function="k_iterm*M*(M-1)*(M-2)")
    reactions = [r1, r2, r3]

    # Define Propagation: P_n + M => P_{n+1}
    for n in range(1, max_state - 1):
        reactions.append(
            gillespy2.Reaction(reactants={f"P{n}": 1, 'M': 1}, products={f"P{n + 1}": 1}, rate="k_p")
        )

    # Define Chain transfer to monomer
    for n in range(1, max_state):
        reactions.append(
            gillespy2.Reaction(reactants={f"P{n}": 1, 'M': 1}, products={f"D{n}": 1, "P1": 1}, rate="k_tm")
        )

    # Define Termination by recombination
    for n in range(1, max_state):
        for m in range(1, max_state):
            nm = n + m
            if(nm < max_state):
                if(n == m):
                    reactions.append(gillespy2.Reaction(
                        reactants={f"P{n}": 1, f"P{m}": 1}, products={f"D{nm}": 1, "P1": 1}, rate="k_tm_2"
                    ))
                else:
                    reactions.append(gillespy2.Reaction(
                        reactants={f"P{n}": 1, f"P{m}": 1}, products={f"D{nm}": 1, "P1": 1}, rate="k_tm"
                    ))
                    
    # Add Reactions to Model
    model.add_reaction(reactions)
    
    # Define Timespace: 0 -> 20 seconds
    tspan = gillespy2.TimeSpan.arange(t=20, increment=0.1)

    # Set Model Timespan
    model.timespan(tspan)
    return model
```

### Instantiate the Model

```python
model = build_styrene_polymerization()
```

***
## Run the Simulations
***

```python
%time result = model.run()
```
